#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import matplotlib.pyplot as plt
```

```python
# View the plot 
%matplotlib inline
```

```python
#cause of deaths
cause=['Chronic Diseases','Unintentional Injury','Alzheimers','Influenza and pneumonia', 'Spesis','Others']

#percentile
percentile =[62,5,4,2,1,26]
```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot properties
plt.figure(figsize=(10,10))
```

```python
#set figure size
```

```python
#explode the largest pie in the data set
explode=(0.05,0,0,0,0,0)
```

```python
#set pie chart properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=80)
#set axis equal to draw pie as circle
plt.axis('equal')
#set title of the pie chart
plt.title('Leading Cause of death')
#show the plot (pie chart)
plt.show()
```
