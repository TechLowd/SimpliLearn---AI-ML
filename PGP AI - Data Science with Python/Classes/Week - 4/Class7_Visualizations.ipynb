### Transformations to handle the skewed distributions

```python
#Fare is a right skewed
#Age is a normal distribution with some outliers - the outliers need to be imputed with some logical value Q3 + 1.5IQR 
# or Q1 - 1.5IQR

#to convert a right skewed distribution to a normal one, you need to try 
#transformations like: sq root, cube root, and incase of extreme right-skewness, try out logarithms
#to convert a left skewed distribution to a normal one, you need to try 
#out the transformations: sq, cube, power 4, and incase of extreme left skewness: try 10pow(x)
```

```python
sns.distplot(np.log10(titanic["Fare"]+1), bins = 50, hist = True, )
```

```python
sns.distplot((titanic["Fare"]**.5), bins = 50, hist = True, )
```

```python
sns.distplot((titanic["Fare"]**(1/3)), bins = 50, hist = True, )
```

```python
sns.distplot((titanic["Fare"]**(1/4)), bins = 50, hist = True, )
```

### Factor + Factor

```python
titanic.groupby("Sex").mean()["Survived"].plot("bar")
```

```python
sns.countplot("Sex", hue = "Survived", data = titanic)
```

```python
#stacked bar plot
pd.crosstab(titanic.Sex, titanic.Survived).plot(kind = "bar", stacked = True)
```

```python
#Beside bar plot
pd.crosstab(titanic.Sex, titanic.Survived).plot(kind = "bar", stacked = False)
```

```python
#Factor plots
import warnings
warnings.filterwarnings(action = "ignore")
sns.factorplot("Embarked", "Survived", data = titanic)
plt.show()
#this also gives the 95% confidence interval for the prob of survival
#With 95% confidence we can say that a person with embarked = S, which prob of survival will be between 30% and 38%
```
