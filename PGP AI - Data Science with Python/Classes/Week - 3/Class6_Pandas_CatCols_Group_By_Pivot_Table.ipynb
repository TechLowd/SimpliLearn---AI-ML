#### Pivot_table command

```python
sales = pd.read_excel("sales-funnel.xlsx")
```

```python
sales.head()
```

```python
s = sales.sort_values(["Manager", "Rep"])
```

```python
s.Status = s.Status.astype("category")
```

```python
s.info()
```

```python
s.Status = s.Status.astype("object")
```

```python
s.info()
```

```python
#to make multiple columns into categorical type
cat_cols = ["Name", "Rep", "Manager", "Status"]
```

```python
for col in cat_cols:
    s[col] = s[col].astype("category")
```

```python
s.info()
```

```python
#simple pivot table
pd.pivot_table(s, index = "Name") #average of each column accordin to the name
```

```python
#simple pivot table
pd.pivot_table(s, index = "Name", values = ["Price", "Quantity"]) #average price and quantity
```

```python
#simple pivot table
pd.pivot_table(s, index = ["Manager", "Rep"], values = ["Price", "Quantity"], aggfunc=np.sum) #average price and quantity
```

```python
#simple pivot table
pd.pivot_table(s, index = ["Manager", "Rep"], values = ["Price", "Quantity"], aggfunc=[np.sum, np.mean, len]) #average price and quantity
```

```python
s.head()
```

```python
#simple pivot table
pd.pivot_table(s, index = ["Manager"], columns = "Product", values = ["Price"], aggfunc=[np.mean], fill_value=0, margins = True)
#margins will give the aggregation row-wise and column wise
```

```python
s[:1]
```

```python
#simple pivot table
pd.pivot_table(s, index = ["Manager"], columns = "Product", values = ["Price", "Quantity"], aggfunc = {"Price": np.mean, "Quantity": np.sum}, 
               fill_value=0, margins = True)
#margins will give the aggregation row-wise and column wise
```
